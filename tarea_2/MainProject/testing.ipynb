```python
encabezado = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "valorant"]
#len(encabezado) : 15
```

```python
df = pd.read_csv("prueba.csv", names=encabezado)
df.head()
```

```python
df.describe()
```

```python
if '?' in df.values:
    # reemplace "?" a NaN
    df.replace("?", np.nan, inplace = True)
elif ' ?' in df.values:
    # reemplace " ?" a NaN
    df.replace(" ?", np.nan, inplace = True)
elif ' ' in df.values:
    # reemplace " ?" a NaN
    df.replace(" ", np.nan, inplace = True)
```

```python
print('?' in df.values)
print(' ?' in df.values)
print(' ' in df.values)

```

```python
# buscando datos nulos

datos_faltantes = df.isnull()
datos_faltantes.head(10)
```

```python
for columna in datos_faltantes.columns.values.tolist():
    print('columna:',columna)
    print (datos_faltantes[columna].value_counts())
    print("")   
```

```python
# para el caso de la columna age se toma la decisi√≥n de calcular su promedio y utilizarlo en las 2 filas
age_pro = df["age"].astype("float").mean(axis=0)
age_pro_r = float(round(age_pro))
print("Promedio de age:", age_pro_r)
```

```python
# reemplazamos 

df["age"].replace(np.nan, age_pro_r, inplace=True)
```
